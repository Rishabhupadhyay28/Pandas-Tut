# APPLY METHOD 

CREATING FUNCTION AND APPLYING IT INTO THE DATA 

```python
def last_number(num):
    return str(num)[-4:]
```

```python
last=data['CC Number'].apply(last_number)
```

```python
data["last number"]=last
```

```python
data
```

```python
def yelp(price):
    if price<10:
        return '$'
    elif price>= 10 and price<30:
        return '$$'
    else:
        return '$$$'
```

```python
data['yelp']=data['total_bill'].apply(yelp)
```

```python
data
```

```python
data["total_bill"].apply(lambda bill:bill*2)
```

```python
data.head()
```

```python
def quality(total_bill,tip):
    if tip/total_bill>0.25:
        return "Generous"
    else:
        return "other"
```

```python
quality(16.99,1.01)
```

```python
data['Ouality']=data[['total_bill','tip']].apply(lambda data:quality(data['total_bill'],data['tip']),axis=1)
```

```python
data
```

```python
import numpy as np 
```

```python
data['Quality']=np.vectorize (quality)(data['total_bill'],data['tip']
                                      )
```

```python
data.head()
```

```python
data=data.drop('Ouality',axis=1)
```

```python
data
```

```python
def tip_pect(tip,total_bill):
    return ((tip/total_bill)*100)
```

```python
data['tip_pect']=data[['total_bill','tip']].apply(lambda data : tip_pect(data['tip'],data['total_bill']),axis=1).round(2)
```

```python
data
```

```python
data["TIP"]=np.vectorize(tip_pect)(data['tip'],data['total_bill']).round(2)
```

```python
data
```
